```python
import pandas as pd
import datetime as datetime
import matplotlib.pyplot as plt
```

```python
accre_df = pd.read_csv('../data/fullsample.csv', usecols= ['JOBID',
                                                         'STATE', 
                                                         'BEGIN', 
                                                         'END', 
                                                         'REQMEM',
                                                         'USEDMEM',
                                                         'PARTITION', 
                                                         'EXITCODE'],
                                                          )
accre_df.sort_values('END')
accre_df.tail(50)
```

```python
#convert begin and end to datetime
#completions per hour would be how many JOBIDs return a STATE of complete per hour.  In other words,
#count how many JOBIDs have STATE completed during the same hour in END
#start by sorting end.
pd.to_datetime(accre_df['END'], 
               #format= '%Y%m%dT%H%I%S', 
               errors = 'coerce')

accre_df.sort_values('END')
accre_df.tail(50)
```

```python
#generate 24 hour moving average for STATE = COMPLETED
rolling = accre_df[['STATE', 'END']]
rolling = rolling[rolling.STATE == 'COMPLETED']
rolling['END'] = pd.to_datetime(rolling['END'],
                                errors = 'coerce')
rolling = (rolling.sort_values('END')
                                     .reset_index()
          )
rolling['Total_Completions_24_Hours'] = (rolling
                             .rolling('24h', on = 'END')['STATE']
                             .count())
rolling['Rolling_Hourly_Average'] = rolling['Total_Completions_24_Hours'] / 24
rolling['Date'] = rolling['END'].dt.date
rolling['raw_count_hour'] = (rolling
                          .rolling('1h', on = 'END')['STATE']
                          .count())
rolling
```

```python
plt.figure(figsize = (12,6))
rolling[rolling['END'] > '2020-11-01 00:00:00'].set_index('END')['Rolling_Hourly_Average'].plot(color = 'red', label = '24_Hour_Rolling_Average')
rolling[rolling['END'] > '2020-11-01 00:00:00'].set_index('END')['raw_count_hour'].plot(label = 'Raw Counts', alpha= 0.5)
plt.title('Daily Counts Completions', fontsize = 18)
plt.legend();
plt.show()
```

```python
# 1 hour moving average for December
rolling['Total_Completions_1_Hour_December'] = (rolling
                             .rolling('1h', on = 'END')['STATE']
                             .count())
rolling['Rolling_1_Minute_Average_December'] = rolling['Total_Completions_1_Hour_December'] / 60
rolling['Total_Completions_5_Minutes'] = (rolling
                             .rolling('5T', on = 'END')['STATE']
                             .count())
rolling['Rolling_5_Minute_Average_December'] = rolling['Total_Completions_5_Minutes'] / 5
rolling['Date'] = rolling['END'].dt.date
rolling['Raw_Count_Minute_December'] = (rolling
                          .rolling('1T', on = 'END')['STATE']
                          .count())
rolling
```
